# Program to display all the header tags from wikipedia.org and make data frame

```python
import  requests
from  bs4 import BeautifulSoup
import pandas as pd

url = requests.get('https://en.wikipedia.org/wiki/Main_Page')
soup = BeautifulSoup(url.content)
soup

header1 = []
for i in soup.find_all('h1', class_='central-textlogo-wrapper'):
    header1.append(i.text)
    header1

    
header2 = []
for i in soup.find_all('h2', class_='mp-h2'):
    header2.append(i.text) 
    
    print(len(header1),len(header2))
    
df=pd.DataFrame({'h2':header2})
df = df.replace('\n',' ', regex=True)
df
```

# python program to display IMDB’s Top rated 50 movies’ data (i.e. name, rating, year of release) and make data frame

```python
imdburl = requests.get('https://www.imdb.com/chart/top/')
imdbsoup = BeautifulSoup(imdburl.content)
imdbsoup

moviename= []
for i in imdbsoup.find_all('td', class_='titleColumn'):
    moviename.append(i.text)
    moviename
    
rating= []
for i in imdbsoup.find_all('td', class_='ratingColumn imdbRating'):
    rating.append(i.text)
    rating
    
yearofrelease= []
for i in imdbsoup.find_all('span', class_='secondaryInfo'):
    yearofrelease.append(i.text)
    yearofrelease
    
moviedf=pd.DataFrame({'Movie Name': moviename[0:50] , 'Rating': rating[0:50], 'Year of Release':yearofrelease[0:50] })
moviedf = moviedf.replace('\n',' ',regex=True)
moviedf
```

# python program to display list of respected former presidents of India(i.e. Name , Term ofoffice) from https://presidentofindia.nic.in/former-presidents.htm and make data frame

```python
poiurl = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
poisoup = BeautifulSoup(poiurl.content)
poisoup

Presidentname= []
for i in poisoup.find_all('div', class_='presidentListing'):
    Presidentname.append(i.text)
    Presidentname
    
termofoffice= []
for i in poisoup.find_all('span', class_='terms'):
    termofoffice.append(i.text)
    termofoffice
    
poidf=pd.DataFrame({'President Name': Presidentname, 'Term of Office': termofoffice})
poidf = poidf.replace('\n',' ', regex=True) 
poidf
```

# python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame

```python
import  requests
from  bs4 import BeautifulSoup
import pandas as pd

newsurl = requests.get('https://www.cnbc.com/world/?region=world')
nsoup = BeautifulSoup(newsurl.content)
nsoup

news = []
for i in nsoup.find_all('a', class_='LatestNews-headline'):
    news.append(i.text.split('|')[0])
    news

newstime = []
for i in nsoup.find_all('time', class_='LatestNews-timestamp'):
    newstime.append(i.text.split('|')[0])
    newstime
    
newslink = []
for i in nsoup.find_all('a', href=True,class_='LatestNews-headline'):
    newslink.append(i['href'])
    newslink
    

newsdf=pd.DataFrame({'Headline': news, 'Time':newstime, 'Newslink':newslink})
newsdf = newsdf.replace('\n',' ', regex=True) 
newsdf
```

# python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles

```python
import  requests
from  bs4 import BeautifulSoup
import pandas as pd

paperurl = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
papersoup = BeautifulSoup(paperurl.content)
papersoup

ptitle= []
for i in papersoup.find_all('h2',class_='sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg'):
    ptitle.append(i.text.split('|')[0])
    ptitle

auth= []
for i in papersoup.find_all('span', class_='sc-1w3fpd7-0 dnCnAO'):
    auth.append(i.text.split('|')[0])
    auth 
    
pdate = []
for i in papersoup.find_all('span',class_='sc-1thf9ly-2 dvggWt'):
    pdate.append(i.text.split('|')[0])
    pdate
    
plink = []
for i in papersoup.find_all('a', href=True, class_='sc-5smygv-0 fIXTHm'):
    plink.append(i['href'])
    plink
    
jdf=pd.DataFrame({'Paper Title': ptitle, 'Author':auth, 'Paper Date': pdate,'Paper Link':plink})
jdf = jdf.replace('\n',' ', regex=True) 
jdf 
```

# python program to scrape mentioned details from dineout.co.in and make data frame

```python
import  requests
from  bs4 import BeautifulSoup
import pandas as pd

dineurl = requests.get('https://www.dineout.co.in/bangalore-restaurants')
dinesoup = BeautifulSoup(dineurl.content)
dinesoup

rest= []
for i in dinesoup.find_all('a', class_='restnt-name ellipsis'):
    rest.append(i.text)
    rest

cuis= []
for i in dinesoup.find_all('span', class_='double-line-ellipsis'):
    cuis.append(i.text.split('|')[1])
    cuis
    
loc= []
for i in dinesoup.find_all('div', class_='restnt-loc ellipsis'):
    loc.append(i.text.split('|')[0])
    loc
    
restrate= []
for i in dinesoup.find_all('span', class_='double-line-ellipsis'):
    restrate.append(i.text.split('|')[0])
    restrate
    
imgurl= []
for i in dinesoup.find_all('img', class_='no-img'):
    imgurl.append(i.get('data-src'))
    imgurl   
    
dinedf=pd.DataFrame({'Restaurant name':rest,'Cuisine':cuis,'Location':loc,'Rate':restrate, 'Image':imgurl })
dinedf = dinedf.replace('\n',' ', regex=True)
dinedf
```
