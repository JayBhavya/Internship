# Q4 Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:Brand,Product Description,Price

```python
import pandas as pd
from selenium import webdriver
import warnings 
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time

driver=webdriver.Chrome(r"C:\Users\HP\Downloads\chromedriver_win32\chromedriver.exe")

driver.get("https://www.flipkart.com/")

lclose=driver.find_element(By.XPATH, '/html/body/div[2]/div/div/button')
lclose.click()


sglass=driver.find_element(By.CLASS_NAME, '_3704LK')
sglass.send_keys('sunglasses')

search=driver.find_element(By.CLASS_NAME, 'L0Z3Pu')
search.click()

driver.get("https://www.flipkart.com/search?q=sunglasses&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")

prod_brand=[]
prod_discr=[]
prod_price=[]

#     brand_tags=driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]')
#     for i in brand_tags[0:10]:
#         brand=i.text
#         prod_brand.append(brand)

        
start=0
end=4

for page in range(start, end):
    
    brand_tags=driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]')
    for i in brand_tags[0:30]:
        prod_brand.append(i.text)
        
    discr_tags=driver.find_elements(By.XPATH, '//a[@class="IRpwTa"]')
    for i in discr_tags[0:30]:
        prod_discr.append(i.text)

    price_tags=driver.find_elements(By.XPATH, '//div[@class="_30jeq3"]')
    for i in price_tags[0:30]:
        prod_price.append(i.text)
        
    nxtbtn=driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
    nxtbtn.click()
    time.sleep(5)
```

```python
import pandas as pd
flipdf=pd.DataFrame({'Brand': prod_brand,'Discription':prod_discr, 'Price':prod_price})
flipdf
```

# Q5: Scrape 100 reviews data from flipkart.com for iphone11 phone. 

```python
import pandas as pd
from selenium import webdriver
import warnings 
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time

driver=webdriver.Chrome(r"C:\Users\HP\Downloads\chromedriver_win32\chromedriver.exe")

driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART")
apprating=[]
appreview=[]
rev_disc=[]
        
start=0
end=10

for page in range(start, end):
    
    ratings_tags=driver.find_elements(By.XPATH, '//div[@class="_3LWZlK _1BLPMq"]')
    for i in ratings_tags:
        apprating.append(i.text)
        
    review_tags=driver.find_elements(By.XPATH, '//p[@class="_2-N8zT"]')
    for i in review_tags:
        appreview.append(i.text)

    rev_tags=driver.find_elements(By.XPATH, '//div[@class="t-ZTKy"]')
    for i in rev_tags:
        rev_disc.append(i.text)
    
    revbtn=driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
    revbtn.click()
    
    time.sleep(11)   
```

```python
import pandas as pd
apppdf=pd.DataFrame({'Rating': apprating,'Review':appreview, 'Detail Discription':rev_disc})
apppdf
```

# Q8: Write a python program to scrape data for Top 1000 Quotes of All Time from https://www.azquotes.com/

```python
import pandas as pd
from selenium import webdriver
import warnings 
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time

driver=webdriver.Chrome(r"C:\Users\HP\Downloads\chromedriver_win32\chromedriver.exe")

driver.get(" https://www.azquotes.com/")

topqts=driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
topqts.click()

quote=[]
author=[]
typeofquote=[]

quote_tags=driver.find_elements(By.XPATH, '//a[@class="title"]')
for i in quote_tags[0:1000]:
    qt=i.text
    quote.append(qt)

author_tags=driver.find_elements(By.XPATH, '//div[@class="author"]')
for i in author_tags[0:1000]:
    auth=i.text
    author.append(auth)

qttype_tags=driver.find_elements(By.XPATH, '//div[@class="tags"]')
for i in qttype_tags[0:1000]:
    tqt=i.text
    typeofquote.append(tqt)
 
```

```python
print(len(quote),len(author), len(typeofquote))
```

```python
import pandas as pd
qtpdf=pd.DataFrame({'Quote':quote,'Author':author,'Type of Quotes':typeofquote})
qtpdf
```
